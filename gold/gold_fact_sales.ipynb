# SQL script to read from sales silver tables and join them into one fact sales table with already made gold customers and products dim tables

```python
query = """
SELECT
    cs.order_number,
    dp.product_key,
    dc.customer_key,
    cs.product_number,
    cs.customer_id,
    cs.order_date,
    cs.ship_date,
    cs.due_date,
    cs.sales_amount,
    cs.quantity,
    cs.price
FROM workspace.silver.crm_sales cs
LEFT JOIN workspace.gold.dim_customers dc
    ON cs.customer_id = dc.customer_id
LEFT JOIN workspace.gold.dim_products dp
    ON cs.product_number = dp.product_number
"""

df = spark.sql(query)
```

```python
df.limit(10).display()
```

# writing to gold table

```python
df.write.mode("overwrite").format("delta").saveAsTable("workspace.gold.fact_sales")
```
